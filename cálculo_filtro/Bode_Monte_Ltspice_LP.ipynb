```python
import ltspice
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator 
from matplotlib.collections import PatchCollection
from matplotlib.patches import Rectangle

# Load the .raw file
l = ltspice.Ltspice('Filtro_norma_monte.raw')
l.parse()

# Get the frequency data
frequency = l.get_data('frequency')
fig_bode, ax_bode = plt.subplots(figsize=(10,4))
fig_phase, ax_phase = plt.subplots(figsize=(10,4))
for i in range(30):
    # Get the voltage data for the input and output nodes
    vout = l.get_data('V(vo_fil)',i)
    # Calculate the transfer function
    transfer_function = vout
    phase = np.angle(transfer_function, deg=True)
    # Plot the Bode diagram for the current simulation
    ax_bode.semilogx(frequency, 20 * np.log10(np.abs(transfer_function)))
    ax_phase.semilogx(frequency, phase)
# Plot the Bode diagram

# Lmites del grfico
ax_bode.set_xlim(70000, 80000)
ax_bode.set_ylim(-3, 1)
ax_bode.xaxis.set_major_locator(MultipleLocator(1000))

# Rotate X-axis tick labels vertically
ax_bode.tick_params(axis='x', rotation=90)

ax_bode.grid(visible=True, which='both', axis='both', linewidth=0.3)
ax_bode.grid(visible=True, which='major', axis='both', linewidth=1)
ax_bode.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
ax_bode.minorticks_on()
ax_bode.add_patch(Rectangle((70000, -1), 5000, 5, linewidth=1, facecolor='green', alpha=0.2))
ax_bode.add_patch(Rectangle((152000, -50), 10000, 10, linewidth=1, facecolor='green', alpha=0.2))
plt.minorticks_on()
plt.legend()
plt.show()
```
